```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat = np.corrcoef(x,y)
    # Return entry [0,1]
    return corr_mat[0,1]
```

Data above suggests that age increase does not affect survival that much, it must be that there are more middle aged people that have survived than older ones.

```python
print(pearson_r(x= titanic_train.age.notnull(), y = titanic_train['survived']))
print(pearson_r(x= titanic_train.pclass, y = titanic_train['survived']))
```

Plot below suggests and confirms that more middle aged people did not survived, as well as that there are more people in class 3 that did not survived,

```python
def permutation_sample(data1, data2):
    """Generate a permutation sample from two data sets."""

    # Concatenate the data sets: data
    data = np.concatenate((data1,data2))

    # Permute the concatenated array: permuted_data
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two: perm_sample_1, perm_sample_2
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]

    return perm_sample_1, perm_sample_2
```

```python
perm_samp_pclass, perm_samp_survived = permutation_sample(titanic_train.pclass, titanic_train.survived)
```

```python
# Compute observed correlation: r_obs
r_obs = pearson_r(titanic_train.pclass, titanic_train.survived)

# Initialize permutation replicates: perm_replicates
perm_replicates = np.empty(100000)

# Draw replicates
for i in range(100000):
    # Permute illiteracy measurments: illiteracy_permuted
    survived_permuted = np.random.permutation(titanic_train.survived)

    # Compute Pearson correlation
    perm_replicates[i] = pearson_r(titanic_train.pclass, survived_permuted)

# Compute p-value: p
p = np.sum(perm_replicates <= r_obs) / len(perm_replicates)

print('p-val =', p)
```

```python
# Load functions
def ecdf(col):
	"""Compute ECDF for a one-dimensional array of measurements."""
	n = len(col) # Number of data points: n
	x = np.sort(col) # x-data for the ECDF: x
	y = np.arange(1, n+1) / float(n) # y-data for the ECDF: y
	return x,y
```

It is seen below that the value of Pearson Correlation coefficient is very low compared to the permutated sample, this suggest that indeed pclass might be negatively correlated to the value of survived.

## Preparing to make the Prediction Model

Now checking for Passenger Title, it can be seen below that some of the passengers have different title, it might needed to be grouped for better analysis.  

```python
titanic_train_age = titanic_train.loc[titanic_train.age.notnull()]
```

```python
_ = pd.scatter_matrix(titanic_train_age.loc[:,['age','fare','pclass']], c = titanic_train_age.survived, figsize = [6,6],
                      s = 150, marker = '.')
plt.show()
```

```python
# Creating lots of Function to change Factors to Numbers 
# as I have not learned how to change it automatically yet.
def gend_no(gender):
    if gender == 'male':
        return 0
    elif gender == 'female':
        return 1
    else:
        return 2

def title_no(title):
    if title == 'Mr':
        return 0
    elif title == 'Mrs':
        return 1
    elif title == 'Miss':
        return 2
    elif title == 'Master':
        return 3
    elif title == 'Honorific_male':
        return 4
    else:
        return 5

def emb_no(embarked):
    if embarked == 'S':
        return 0
    elif embarked == 'C':
        return 1
    else:
        return 2    
```

```python
titanic_train_age['gender_no'] = titanic_train_age.gender.apply(gend_no)
titanic_train_age['title_no'] = titanic_train_age.title.apply(title_no)
titanic_train_age['embarked_no'] = titanic_train_age.embarked.apply(title_no)
```

```python
print(titanic_train_age.gender_no.value_counts())
print(titanic_train_age.title_no.value_counts())
print(titanic_train_age.embarked_no.value_counts())
```

```python
titanic_test_age = titanic_test.loc[titanic_test.age.notnull()]
```

```python
titanic_test_age['gender_no'] = titanic_test_age.gender.apply(gend_no)
titanic_test_age['title_no'] = titanic_test_age.title.apply(title_no)
titanic_test_age['embarked_no'] = titanic_test_age.embarked.apply(emb_no)
```

```python
print(titanic_test_age.gender_no.value_counts())
print(titanic_test_age.title_no.value_counts())
print(titanic_test_age.embarked_no.value_counts())
```

```python
titanic_test_age = titanic_test_age.sort_values(by='name')
titanic_test_age = titanic_test_age.reset_index()
```

## Creating the actual prediction model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 8)
```

```python
train_X = titanic_train_age.loc[:,['age','fare','pclass','parch','sibsp','gender_no','title_no','embarked_no']]
train_y = titanic_train_age.survived
knn.fit(train_X, train_y)
```

```python
test_X = titanic_test_age.loc[:,['age','fare','pclass','parch','sibsp','gender_no','title_no','embarked_no']]
predict = knn.predict(test_X)
```

```python
titanic_test_age.loc[:,'survived'] = predict
```
